```python
#drop rows with missing values
df = df.dropna()
df.isnull().sum()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
type(df['InvoiceDate'][0])
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Date-related features
df['InvoiceYear'] = df['InvoiceDate'].dt.year
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
df['InvoiceDay'] = df['InvoiceDate'].dt.day
df['InvoiceHour'] = df['InvoiceDate'].dt.hour

# Optional: Day of week
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
```

```python
df
```

```python
customer_df = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',         # Unique invoices per customer
    'Quantity': 'sum',              # Total quantity purchased
    'TotalPrice': 'sum',            # Total money spent
    'InvoiceDate': ['min', 'max']   # First and last purchase
}).reset_index()

# Rename columns
customer_df.columns = [
    'CustomerID', 'NumOrders', 'TotalQuantity', 'TotalSpent',
    'FirstPurchaseDate', 'LastPurchaseDate'
]

# Calculate recency (how many days since last purchase)
latest_date = df['InvoiceDate'].max()
customer_df['RecencyDays'] = (latest_date - customer_df['LastPurchaseDate']).dt.days
```

```python
customer_df
```

```python
df.to_csv('cleaned_transactions.csv', index=False)
customer_df.to_csv('customer_summary.csv', index=False)
```
